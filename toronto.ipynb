```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header=0, flavor='html5lib')[0]
```

```python
#### Dropping any row that has a borough that's 'Not Assigned'
df.drop(df[df['Borough'] == 'Not assigned'].index, 0, inplace = True) 
```

```python
#### Replacing ant neighbourhood that's 'Not assigned' to the borough 
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df['Borough']
```
