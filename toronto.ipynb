```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header=0, flavor='html5lib')[0]
```

```python
#### Dropping any row that has a borough that's 'Not Assigned'
df.drop(df[df['Borough'] == 'Not assigned'].index, 0, inplace = True) 
```

```python
#### Replacing ant neighbourhood that's 'Not assigned' to the borough 
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df['Borough']
```

```python
### Getting latitude and longitude from the CSV file 
latlong = pd.read_csv('https://cocl.us/Geospatial_data')
latlong.head()
```

```python
# Mergining the imported list with the created list from the wikipedia page
neighborhoods = pd.merge(df, latlong, 'left', 'Postal Code')
neighborhoods.head()
```

# Mapping neighbourhoods

```python
### Getting center points for Toronto to start map off
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map, labeling by bourogh name and not neighborhood 
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{} - {}'.format(borough, neighborhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
